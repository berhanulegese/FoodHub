### **Question 13:** The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of more than 50 and the average rating should be greater than 4. Find the restaurants fulfilling the criteria to get the promotional offer. [3 marks]

```python
# Write the code here
# Convert rating to numeric, coercing errors (like 's') to NaN
df["rating"] = pd.to_numeric(df["rating"], errors='coerce')

# Drop rows where rating is NaN
df_clean = df.dropna(subset=["rating"])

# Group and filter eligible restaurants
rating_summary = df_clean.groupby("restaurant_name")["rating"].agg(["count", "mean"]).reset_index()

eligible_restaurants = rating_summary[
    (rating_summary["count"] > 50) &
    (rating_summary["mean"] > 4)
].rename(columns={"count": "rating_count", "mean": "average_rating"})

# View result
print(eligible_restaurants)
```

#### Observations:
#### Blue Ribbon Fried Chicken has 64 ratings and a rating of 4.28125

####  Blue Ribbon Sushi has  73 ratings and a rating of 4.219178

#### The Meatball Shop has 84 ratings and a rating of 4.511905

### **Question 14:** The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders. [3 marks]

```python
# Write the code here

#calculate revenue
def calculate_revenue(x):
    if x > 20:
        return x*0.25
    elif x > 5:
        return x*0.15
    else:
        return x*0

#apply the 25 and 15 % on all orders cost and store it in new column revenue
df['Revenue'] = df['cost_of_the_order'].apply(calculate_revenue)
df.head()

total_revenue = df['Revenue'].sum() #sum all revenues
print('The net revenue is around', round(total_revenue, 2), 'dollars')
```

#### Observation
#### The net revenue is around 6166.3 dollars

### **Question 15:** The company wants to analyze the total time required to deliver the food. What percentage of orders take more than 60 minutes to get delivered from the time the order is placed? (The food has to be prepared and then delivered.) [2 marks]

```python
# Write the code here
# calculate total tike taken from food preparation and delivery time and store in in a new column named total_time
df['total_time'] = df['food_preparation_time'] + df['delivery_time']

#orders that take more than 60 min
orders_more_than_60_min = df[df['total_time'] > 60]

#calculate percentage
percentage = (len(orders_more_than_60_min) / len(df)) * 100
print(round(percentage,0), '%')
```

 ### **Question 17:** What are your conclusions from the analysis? What recommendations would you like to share to help improve the business? (You can use cuisine type and feedback ratings to drive your business recommendations.) [6 marks]

### Conclusions:

1. **Delivery Time Plays a Role in Customer Satisfaction**
   - Lower-rated orders (e.g., 3 stars) took longer to deliver.
   - Weekends (22 minutes) saw shorter delivery times than weekdays (28 minutes).

2. **Type of Cuisine Influences Delivery Experience**
   - Mexican, Spanish, and Vietnamese cuisine saw more variation in preparation and delivery time.
- Korean and Vietnamese took shorter, consistent times—testifying to smooth operations.

3. **Higher Cost Orders Come with Better Ratings**
   - 5-star rated orders had higher mean cost (~$17), suggesting better perception of value.
   - Lower-cost orders got more diverse feedback.

4. **No Good Correlation Between Variables**
- Cost, preparation time, and delivery time are almost uncorrelated (close to zero correlation measures), and thus must be controlled independently.

5. **11% of Orders Take Over 60 Minutes**
   - These are sources of potential dissatisfaction and must be flagged or examined.

6. **Few Restaurants Are Promotable**
   - Not many restaurants met on both the high-rating and high-volume bases for being eligible for promotion offers.



### Recommendations

1. **Enhance Weekday Delivery Times**
   - Reduce the ~28-minute weekday mean by:
     - Enhancing route planning
     - Encourage order fulfillment off-peak hours
     - Additional weekday delivery staff

2. **Promote High-Rated, High-Volume Restaurants**
   - Promote only restaurants with:
     - Rating count > 50
     - Average rating > 4.0

3. **Operational Improvements Based on Cuisine**
   - Focus on cuisines with inconsistent service (Spanish, Mexican) by:
     - Streamlining kitchen operations
     - Staff training to ensure consistency

4. **Encourage Higher-Value Orders**
   - Offer bundles or upsells to drive order value, which is positively correlated with higher ratings.

5. **Investigate Extended Delivery Times**
- Identify and resolve issues for orders taking longer than 60 minutes.
- Explore issuing compensations to affected clients or rescheduling vendor SLAs.

6. **Boost Customer Rating Involvement**
- Leverage post-delivery reminders or offer incentives (e.g., loyalty points) to encourage more users to rate.
